- Handling missing values 

| Missing                   | % Missing | Action                                                |
| ------------------------- | --------- | ----------------------------------------------------- |
| `Tot_Benes` = 21,596      | \~14%     | **Fill with 0** or **median**        |
| `GE65_Tot_*` ≈ 21,000     | \~14%     | **Fill with 0** (common for suppressed Medicare data) |
| `GE65_Tot_Benes` = 48,848 | \~33%     | Consider **dropping this column** (too sparse)        |
| `Prscrbr_Geo_Cd` = 3,608  | \~2.4%    | Likely "National" → **Keep as is**                    |
| `Prscrbr_Geo_Desc` = 1    | \~0%      | **Drop the row** or **fill**                          |


```python
# Drop GE65_Tot_Benes (too many missing)
data = data.drop(columns=['GE65_Tot_Benes'])

# Fill remaining GE65_* and Tot_Benes with 0
cols_to_fill = ['Tot_Benes', 'GE65_Tot_Clms', 'GE65_Tot_30day_Fills', 'GE65_Tot_Drug_Cst']
data[cols_to_fill] = data[cols_to_fill].fillna(0)

# Drop the one row with missing Geo_Desc
data = data.dropna(subset=['Prscrbr_Geo_Desc'])
```
